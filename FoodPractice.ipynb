```python
## for parsing the metadata

def parse(path):
    g = gzip.open(path, 'r')
    for l in g:
        yield eval(l)
```

```python
##reading the meta dataset
metaData = []
for i in parse(path+'/dataset/meta_Baby.json.gz'):
    metaData.append(i)
metaData
```

```python
### turning the metadata into a dataframe to make it more readable
metaDataset = pd.DataFrame(metaData)
metaDataset.head()
```
